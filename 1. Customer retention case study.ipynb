### Removing unncessary syntax

```python
from string import digits
rem_digits = str.maketrans('', '', digits)
data.columns = data.columns.str.replace('\t','')
data.columns = data.columns.str.translate(rem_digits)
data.columns = data.columns.str.strip()
```

### Now let's visualize the distribution of data to have a better idea on how important are the features

```python
data_cat = data.drop(columns = ['What is the Pin Code of where you shop online from?'], axis = 1)

for i in data_cat:
    if i!='What is the Pin Code of where you shop online from?':
        plt.figure(figsize=(8,6))
        data[i].value_counts().plot.pie(autopct='%1.1f%%')
        plt.xlabel(i, fontsize = 15)
        plt.ylabel('')
        plt.figure()
```
